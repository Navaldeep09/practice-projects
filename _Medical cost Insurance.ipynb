```python
plt.figure(figsize=(22,10))
sns.heatmap(df.describe(),annot=True,fmt='0.2f',linewidth=0.2,linecolor='black',cmap='Spectral')
plt.xlabel('Figure',fontsize=14)
plt.ylabel('Features Name',fontsize=14)
plt.title('Descriptive Graph',fontsize=20)
plt.show()
```

```python
sns.pairplot(data=df,palette="Dark2")
```

```python
#Checking the outliers
plt.figure(figsize=(20,25))
p=1
for i in df.columns:
    if p<=13:
        plt.subplot(5,4,p)
        sns.boxplot(data=df,y=i,palette="Set2_r")
        plt.xlabel(i)
    p+=1
plt.show()
```

```python
from scipy.stats import zscore

# Selecting only numerical features
numeric_features = ['age', 'bmi', 'children', 'charges']
numeric_data = df[numeric_features]

# Calculating Z-scores for numerical features
z_scores = np.abs(zscore(numeric_data))
z_scores
```
